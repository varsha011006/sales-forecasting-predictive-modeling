```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
X = df[['Month']]
y = df['Sales']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
future_months = pd.DataFrame({
    'Month':[13,14,15,16]
})

predictions = model.predict(future_months)

print(predictions)
```

```python
plt.scatter(df['Month'],df['Sales'])

plt.plot(
    future_months['Month'],
    predictions
)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales Forecasting")

plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
pred = model.predict(X)

print(
    r2_score(y,pred)
)
```

```python
plt.scatter(df['Month'],df['Sales'])

plt.plot(
    future_months['Month'],
    predictions
)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales Forecasting")

plt.savefig(
    'sales_forecast.png'
)

plt.show()
```
